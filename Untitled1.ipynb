```python
import pandas as pd
```

```python
df=pd.read_csv("IRIS.csv")
df.head(10)
df.shape
df.dtypes
 
```

```python
#petal length>4.5
df[df["petal_length"]>4.5]
```

```python
#species="Iris-virginica"
df[df["species"]=="Iris-virginica"]
```

```python
#petal average length
df.groupby("species")["sepal_length"].mean()
```

```python
#maximum petal_width
df.groupby("species")["petal_width"].max()
```

```python
#standard deviation of sepal_width
df.groupby("species")["sepal_width"].std()
```

```python
#createanewcolumn:D."petal_ratio"=petal_length/petal_width•Foreachspecies,findtheaveragepetal_ratio
df["petal_ratio"]=df["petal_length"]/df["petal_width"]
df
df.groupby("species")["petal_ratio"].mean()
```

```python
df.to_csv("iris")
```
