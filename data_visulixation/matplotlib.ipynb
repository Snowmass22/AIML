```python
days=["mon","tues","wed","thurs","fri"]
cities=["new york","london","delhi","tokyo"]
temp=[
    [22,23,21,24,25],
    [18,19,17,20,21],
    [30,32,31,33,34],
    [25,26,24,27,28]
]
city=0
fig,axes=plt.subplots(2,2)
for i in range(2):
    for j in range(2):
        axes[i][j].plot(days,temp[city])
        axes[i][j].set_title(cities[city])
        axes[i][j].grid(True)
        fig.suptitle("temperatures in ities")
        city=city+1
```
