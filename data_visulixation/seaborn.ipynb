```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
tips=sns.load_dataset("tips")
print(type(tips))
print(tips.head())
```

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker"
)
```

```python
#relational plots
#scatter plot
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)

#line plot
sns.lineplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
#line plot
flights=sns.load_dataset("flights")
sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)
```

```python
#categorial plots(bar , box)
#bar plot
sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
#box plot
sns.boxplot(
    data=tips,
    x="day",
    y="tip"
)
```

```python
#distribution plots
penguin=sns.load_dataset("penguins")
penguin.head()

sns.histplot(
    data=penguin,
    x="body_mass_g",
    bins=20
)
```

```python
#matrix plot
import pandas as pd
flights=flights.pivot(index="month",columns="year",values="passengers")

sns.heatmap(
    data=flights,
    cmap="coolwarm"
)
```

```python
fig,ax=plt.subplots()

sns.lineplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    ax=ax
)
```
