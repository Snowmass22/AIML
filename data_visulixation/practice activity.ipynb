```python
penguins=sns.load_dataset("penguins")
penguins.head()
sns.lineplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g"
)
plt.show()
```

```python
tips=sns.load_dataset("tips")
tips.head()
sns.boxplot(
    data=tips,
    x="total_bill",
    y="day",
    hue="sex"
)
```

```python
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker"
)
plt.title("total_bill vs tips")
```
