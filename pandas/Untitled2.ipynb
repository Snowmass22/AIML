```python
import requests
from bs4 import BeautifulSoup
```

```python

url="https://quotes.toscrape.com/page/1/"
page_count=1

while True:
    url=f"https://quotes.toscrape.com/page/{page_count}/"
    res=requests.get(url)
    print(res)
    soup=BeautifulSoup(res.text,"lxml")
    quotes=soup.select("div.quote")
    if not quotes:
        print("not valid pages anyomore")
        break
    with open(f"data/quotes{page_count}.html","w",encoding="utf-8") as f:
        f.write(res.text)
        print(f"download complete for page{page_count}")
    page_count=page_count+1
 
```

```python
while True:
    all_quotes=soup.select("div.quote")
    life_quotes=[]
    for q in all_quotes:
        all_tags=[]

        for tag in q.select(".tags .tag"):
            all_tags.append(tag.get_text())

        if "life" in all_tags:
        text=q.select_one("span.text").get_text()
        author=q.select_one("small.author").get_text()
        life_quotes.append([text,author])
print(life_quotes)
```
